```python
target= {'FAKE': 0, 'REAL': 1}
df['label_num']= df['label'].map(target)

df.head()
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df.head()
```

```python
import spacy
nlp=spacy.load('en_core_web_lg')
```

Understanding how spacy works with word embeddings

```python
doc= nlp('Top trump surrogate brutally stabs him')
doc.vector
```

```python
df['vectorized_text']=df.text.apply(lambda x: nlp(x).vector)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test = train_test_split(
df.vectorized_text.values,
    df.label_num,
    test_size= 0.2,
    stratify= df.label_num
)
```

```python
X_train
```

Because the we can only work with a 2D array so we need to stack X_train and X_test

```python
import numpy as np

X_train_2d= np.stack(X_train)
X_test_2d= np.stack(X_test)
```

```python
X_train_2d
```

I can't still use this stacked 2d arrrays, but the machine learning models take negative values as input , I still need to scale the values.

```python
from sklearn.preprocessing import MinMaxScaler


scaler= MinMaxScaler()
scaled_X_train_2d=scaler.fit_transform(X_train_2d)
scaler_X_test_2d=scaler.transform(X_test_2d)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

model = MultinomialNB()

model.fit(scaled_X_train_2d, y_train)

y_preds= model.predict(scaler_X_test_2d)

print(classification_report(y_test,y_preds))
```

Was not impressed with the answer that I got . Let me try KNeighbours 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report 


model = KNeighborsClassifier(n_neighbors= 5, metric='euclidean')

model.fit(scaled_X_train_2d, y_train)

y_preds= model.predict(scaler_X_test_2d)

print(classification_report(y_test,y_preds))
```
